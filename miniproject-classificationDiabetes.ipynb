```python
# Import library
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import accuracy_score
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset Diabetes
data = load_diabetes()

# Menentukan fitur(X) dan target(y)
X = data.data
y = (data.target > data.target.mean()).astype(int) #tipe boolean dengan 1 = diabetes; dan 0 = non

# Menampilkan informasi dataset
print("Nama Fitur:", data.feature_names)
print("Shape Data:", X.shape)
print("Contoh Fitur:\n", pd.DataFrame(X, columns=data.feature_names).head())
print("\nTarget (diabetes or not):\n", pd.Series(y).value_counts())
```

```python
# Bagi dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Buat model Decision Tree Classifier
model = DecisionTreeClassifier(random_state=42, max_depth=3)

# Latih model
model.fit(X_train, y_train)
```

```python
# Prediksi data uji
y_pred = model.predict(X_test)

# Menghitung akurasi model
accuracy = accuracy_score(y_test, y_pred)
print(f"\nAkurasi Model: {accuracy * 100:.2f}%")

# Visualisasi Decision Tree
plt.figure(figsize=(15, 10))
plot_tree(model, feature_names=data.feature_names, class_names=["No Diabetes", "Diabetes"], filled=True)
plt.title("Decision Tree Visualization")
plt.show()
```

```python
# Membuat prediksi
sample = [X_test[20]]  # Contoh Ambil satu sampel dari data uji
prediction = model.predict(sample)
print(f"\nPrediksi untuk data sampel: {'Diabetes' if prediction[0] == 1 else 'No Diabetes'}")
```
